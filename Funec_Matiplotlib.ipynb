```python
ApplePhone = [8.3, 9.6, 20.9, 18.1, 5.0, 24.9, 35.2, 30.1, 10.7, 8.7, 8.3, 13.1]
XionesPhone = [4.1, 4.2, 10.3, 10.0, 3.6, 14.7 , 30.2, 10.9, 6.8, 2.3, 4.2, 12.7] 
BikePhone = [1.0, 4.5, 7.8, 3.7, 7.8, 20.0, 39.0, 12.0, 14.5, 3.7, 2.2, 1.1]
```

```python
meses = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho", "Julho", "Agosto", "Setembro", 
        "Outubro", "Novembro", "Dezembro"]
```

```python
plt.figure(figsize = (20,10))
plt.plot(meses, ApplePhone, color = "red", label = "Gastos em milhares de reais por mês com o ApplePhone")
plt.plot(meses, XionesPhone, color = "yellow", 
         label = "Gastos em milhares de reais por mês com o XionesPhone")
plt.plot(meses, BikePhone, color = "gray", 
         label = "Gastos em milhares de reais por mês com o BikecholaPhone")
plt.title("Gastos da empresa Viciocell em suas marcas de celulares durante o ano")
plt.grid(True)
plt.box(False)
plt.legend()
plt.xlabel("Meses do ano")
plt.ylabel("Gastos em milhares de reais")
```

```python
from numpy import linspace, sin, cos
```

```python
plt.figure(figsize = (10,5))
valoresX = linspace(-0.10, 0.10)
plt.plot(valoresX, (valoresX ** 2) * sin(1 / valoresX), color = "Red", label = "f(x)=x²sen(1/x)")
plt.plot(valoresX, (valoresX ** 2) * cos(1 / valoresX), color = "Red", label = "f(x)=x²cos(1/x)")
plt.title("Plotando um gráfico através de uma função")
plt.grid(True)
plt.box(False)
plt.legend()
```

```python
dias = ["Segunda-feira", "Terça-feira", "Quarta-feira", "Quinta-feira", "Sexta-feira", "Sábado"]
horas = [3.4, 9.0, 9.0, 9.0, 8.40, 1.0]
plt.figure(figsize = (15,10))
plt.bar(dias, horas, color = "Blue", label = "Horas de estudo / Dia")
plt.grid(False)
plt.box(False)
plt.title("Horas de estudo/Dia Engenharia de controle e Automação ")
plt.legend()
```

```python

plt.figure(figsize = (15,10))
plt.pie(horas, labels = dias, autopct = '%1.1f%%', shadow = True, startangle = 90)
plt.title("Porcentagem de Horas de estudo/Engenharia de controle e Automação ")
```
